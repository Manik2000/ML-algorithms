```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs, make_moons

from clustering.k_means import KMeans
```

```python
sns.set_style('whitegrid')
```

## Data

```python
blobs_x, blobs_y = make_blobs(n_samples=500, centers=3, n_features=2, random_state=0, cluster_std=0.6)
```

```python
centroids = blobs_x[np.random.choice(blobs_x.shape[0], 3, replace=False)]
```

```python
sns.scatterplot(x=blobs_x[:, 0], y=blobs_x[:, 1], hue=blobs_y, palette='deep')
plt.show()
```

```python
moons_x, moons_y = make_moons(n_samples=500, noise=0.1, random_state=0)
```

```python
sns.scatterplot(x=moons_x[:, 0], y=moons_x[:, 1], hue=moons_y, palette='deep')
plt.show()
```

## Clustering

```python
k_means_blobs = KMeans(n_clusters=3)
```

```python
def create_dataset_for_plotting(x, y, cluster_labels):
    return pd.DataFrame({'x1': x[:, 0], 'x2': x[:, 1], 'y': y, 'cluster': cluster_labels})
```

```python
np.random.seed(0)
blobs_clusters_1 = k_means_blobs.fit_predict(blobs_x)
```

```python
df_blobs_1 = create_dataset_for_plotting(blobs_x, blobs_y, blobs_clusters_1)
```

```python
sns.scatterplot(x='x1', y='x2', hue='cluster', style="y", data=df_blobs_1, palette='deep')
plt.show()
```

```python
np.random.seed(4002)
blobs_clusters_2 = k_means_blobs.fit_predict(blobs_x)
```

```python
df_blobs_2 = create_dataset_for_plotting(blobs_x, blobs_y, blobs_clusters_2)
```

```python
sns.scatterplot(x='x1', y='x2', hue='cluster', style="y", data=df_blobs_2, palette='deep')
plt.show()
```

```python
k_means_moons = KMeans(n_clusters=2)
```

```python
np.random.seed(0)
moons_clusters_1 = k_means_moons.fit_predict(moons_x)
```

```python
df_moons_1 = create_dataset_for_plotting(moons_x, moons_y, moons_clusters_1)
```

```python
sns.scatterplot(x='x1', y='x2', hue='cluster', style="y", data=df_moons_1, palette='deep')
plt.show()
```
